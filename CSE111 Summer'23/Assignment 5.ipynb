```python
#Task 1

class Book:
    def __init__(self, name, author, genre):
        self.__title = name
        self.__author = author
        self.__genre = genre
        self.__available = True
        self.__borrower = None

    def getTitle(self):
        return self.__title
    
    def setTitle(self, title):
        self.__title = title

    def getAuthor(self):
        return self.__author
    
    def setAuthor(self, author):
        self.__author = author

    def getGenre(self):
        return self.__genre
    
    def setGenre(self, genre):
        self.__genre = genre

    def getAvailable(self):
        return self.__available

    def setAvailable(self, available):
        self.__available = available

    def getBorrower(self):
        return self.__borrower

    def setBorrower(self, borrower):
        self.__borrower = borrower   

    def display_info(self):
        print(f"Title: {self.__title}")
        print(f"Author: {self.__author}")
        print(f"Genre: {self.__genre}")
        if self.__available:
            print("Availability: Available")

        else:
            print("Availability: Not Available")
            
        print("---------------")


class LibraryMember:
    def __init__(self, id, name):
        self.__member_id = id
        self.__name = name
        self.__borrowed_books = []

    def getId(self):
        return self.__member_id
    
    def setId(self, newid):
        self.__member_id = newid
    
    def getName(self):
        return self.__name

    def setName(self, name):
        self.__name = name   

    def borrow_book(self, book):
        if book.getAvailable():
            self.__borrowed_books.append(book)
            book.setBorrower(self.__member_id)
            book.setAvailable(False)
            print(f'{book.getTitle()} is being boorrowed by {self.getName()}.')

        else:
            print("This book is not available at the moment.")

    def return_book(self, book):
        if book in self.__borrowed_books:
            self.__borrowed_books.remove(book)
            book.setBorrower(None)
            book.setAvailable(True)
            print(f'{book.getTitle()} is being returned by {self.getName()}.')
        
        else:
            print(f"{self.getId()} never borrowed this book.")

    def display_borrowed_books(self):
        print(f'List of books borrowed by {self.__name}: ')
        print()
        for i in self.__borrowed_books:
            i.display_info()


class Library:
    def __init__(self):
        self.__books_available = []
        self.__library_member = []

    def add_book(self, book):
        self.__books_available.append(book)
        print(f'{book.getTitle()} has been added to the Library Collection')

    def add_library_member(self, member):
        self.__library_member.append(member)
        print(f'Welcome to the library, {member.getName()}')

    def display_book_list(self):
        print("All the in the Library Collection: ")
        print()
        for i in self.__books_available:
            i.display_info()

    def display_library_members(self):
        print("All the existing members of the Library: ")
        print()
        for i in self.__library_member:
            print(f'Name: {i.getName()}')
            print(f'Member ID: {i.getId()}')
            print("---------------")


# Creating a Library
library = Library()
print("====================================================================================")
# Creating Books
book1 = Book("Harry Potter and the Philosopher's Stone", "J.K. Rowling", "Fiction")
book2 = Book("Harry Potter and the Prisoner of Azkaban", "J.K. Rowling", "Fiction")
book3 = Book("Harry Potter and the Chamber of Secrets", "J.K. Rowling", "Fiction")
book4 = Book("Harry Potter and the Goblet of Fire", "J.K. Rowling", "Fiction")
book5 = Book("Harry Potter and the Order of the Phoenix", "J.K. Rowling", "Fiction")
print("====================================================================================")

# Adding Books to the Library
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.add_book(book4)
library.add_book(book5)
print("====================================================================================")

# Creating Members
member1 = LibraryMember("M1", "Lewis Hamilton")
member2 = LibraryMember("M2", "Sebestian Vettel")
member3 = LibraryMember("M3", "Sergio Perez")
print("====================================================================================")

# Adding members to the Library
library.add_library_member(member1)
library.add_library_member(member2)
library.add_library_member(member3)
print("====================================================================================")

#Displaying all the books in the Library
library.display_book_list()
print("====================================================================================")

#Displaying all the members of the Library
library.display_library_members()
print("====================================================================================")

#Borrow Book
member1.borrow_book(book2)
member3.borrow_book(book5)
member2.borrow_book(book2)
member1.borrow_book(book3)
member2.borrow_book(book1)
print("====================================================================================")

#Displaying Books borrowed by the members
member1.display_borrowed_books()
print("====================================================================================")

member2.display_borrowed_books()
print("====================================================================================")

member3.display_borrowed_books()
print("====================================================================================")

#Return Book
member3.return_book(book5)
member1.return_book(book2)
member2.return_book(book4)
print("====================================================================================")

#Displayig all the books in the Libray again.
library.display_book_list()
print("====================================================================================")
```
