```python
#Task 1

class Fruit:
    def __init__(self, formalin=False, name=''):
        self.__formalin = formalin
        self.name = name
    
    def getName(self):
        return self.name
    
    def hasFormalin(self):
        return self.__formalin
    

class testFruit:
    def test(self, f):
        print('----Printing Detail----')
        if f.hasFormalin():
            print('Do not eat the',f.getName(),'.')
            print(f)

        else:
            print('Eat the',f.getName(),'.')
            print(f)
    

class Mango(Fruit):
    def __init__(self, formaline = True, name = "Mango"):
        super().__init__(formaline, name)

    def __str__(self):
        if self.hasFormalin():
            return f'{self.name}s are bad for you'

        else:
            return f'{self.name}s are good for you'


class Jackfruit(Fruit):
    def __init__(self, formaline = False, name = "Jackfruit"):
        super().__init__(formaline, name)

    def __str__(self):
        if self.hasFormalin():
            return f'{self.name}s are bad for you'

        else:
            return f'{self.name}s are good for you'

    
m = Mango()
j = Jackfruit()
t1 = testFruit()
t1.test(m)
t1.test(j)
```

```python
#Task 2

class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
        
    def examSyllabus(self):
        return "Maths , English"
    
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"
    

class ScienceExam(Exam):
    def __init__(self, marks, time, *parts):
        super().__init__(marks)
        self.time = time
        self.syllabus = ["Maths", "English"]
        self.number_of_parts = 2
        for i in parts:
            self.number_of_parts += 1
            self.syllabus.append(i) 

    def examSyllabus(self):
        return ", ".join(self.syllabus)
    
    def examParts(self):
        num = 1
        output = ""
        for i in self.syllabus:
            output += f'Part {num} - {i}'
            output += "\n"
        
        return output
        
    def __str__(self):
        return f'Marks: {self.marks} Time: {self.time} minutes Number of Parts: {self.number_of_parts}'   


engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================') 
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#Task 3

class PokemonBasic:
    def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type

    def get_type(self):
        return 'Main type: ' + self.type

    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'

    def __str__(self):
        return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness


class PokemonExtra(PokemonBasic):
    def __init__(self, name = "Default", hp = 0, weakness = "None", type = "Unknown", second_type = "Unknown", other_move = "Unknown"):
        super().__init__(name, hp, weakness, type)
        self.second_type = second_type
        self.other_move = other_move

    def get_type(self):
        if self.second_type == "Unknown":
            return super().get_type()
        
        else:
            return f'{super().get_type()}, Secondary type: {self.second_type}'

    def get_move(self):
        if self.other_move == "Unknown":
            return super().get_type()
        
        else:
            return f'{super().get_move()}\nOther move: {", ".join(self.other_move)}'


print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())

print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```
