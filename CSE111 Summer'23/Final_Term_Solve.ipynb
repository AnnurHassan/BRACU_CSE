```python
# Question 1

class Monster:
    monsterCount = 0

    def __init__(self, name, power):
        self.name = name 
        self.power = power
        self.alive = True
        Monster.monsterCount += 1

    def get_details(self):
        return (f'Name: {self.name}\nPower: {self.power}\nAlive: {self.alive}')

    def attack(self, *monsters):
        if len(monsters) == 0:
            print("No monster to attack.")
        
        else:
            for i in monsters:
                if self.alive != True:
                    print(f'{self.name} is not alive to attack.')
                
                else:
                    if i.alive != True:
                        print(f'Cannot attack {i.name}. It is not alive. Attack unsuccessful.')
                    
                    else:
                        if self.power > i.power:
                            print(f'Attack successful.{i.name} was defeated by {self.name}')
                            i.alive = False
                            Monster.monsterCount -= 1
                        
                        else:
                            print(f'{self.name} was defeated by {i.name}')
                            self.alive = False
                            Monster.monsterCount -= 1


monster1 = Monster('Godzilla', 40)
monster2 = Monster('Hydra', 30)
monster3 = Monster('KingKong', 50)
print(f"Number of monsters alive:{Monster.monsterCount}")
print('1--------------------------------')
print(monster1.get_details())
print('2--------------------------------')
print(monster2.get_details())
print('3--------------------------------')
print(monster3.get_details())
print('4--------------------------------')
monster1.attack()
print('5--------------------------------')
monster1.attack(monster2)
print('6--------------------------------')
monster1.attack(monster2, monster3)
print('7--------------------------------')
print(f"Number of monsters alive:{Monster.monsterCount}")
print('8--------------------------------')
print(monster2.get_details())
print('9--------------------------------')
monster2.attack(monster1)
```

```python
# Question 2

class User:
    activities = ["Post", "Like", "Comment"]

    def __init__(self, name,email):
        self.name = name
        self.email = email

    def UserActivity(self, activityType):
        if activityType in User.activities:
            return True

        else:
            return False

    def userDetail(self):
        return f"User Detail:\nName:{self.name}\nEmail: {self.email}"
    

class BracbookUser(User):
    def __init__(self, name, email, phn = "Not set"):
        super().__init__(name, email)
        self.phone = phn
        self.activity_log = []

    def userDetail(self):
        if self.activity_log:
            return f'{super().userDetail()}\nPhone: {self.phone}\nActivity Log: {", ".join(self.activity_log)}'
        
        else:
            return f'{super().userDetail()}\nPhone: {self.phone}\nActivity Log: No recent activity.'
            
    
    def UserActivity(self, activityType):
        if super().UserActivity(activityType):
            print(f'{self.name} has {activityType}(d/ed) successfully')
            self.activity_log.append(activityType)

        else:
            print(f"No activities found like {activityType}") 


user1 = BracbookUser("Rakait","xyz@gmail.com")
print("1===========================")
print(user1.userDetail())
print("2===========================")
user2 = BracbookUser("Sazzad","abc@gmail.com",
"01727xxxxxx")
print("3===========================")
print(user2.userDetail())
print("4===========================")
user1.UserActivity("Like")
print("5===========================")
user1.UserActivity("Comment")
print("6===========================")
print(user1.userDetail())
print("7===========================")
user2.UserActivity("Share")
print("8===========================")
user2.UserActivity("Comment")
print("9===========================")
print(user2.userDetail())
```

```python
# Question 1

class Netflix:
    total_shows = 0
    show_list = []

    def __init__(self, name, genre, ep = 10):
        self.show_name = name
        self.no_of_episodes = ep
        self.genre = genre
        Netflix.total_shows += 1
        Netflix.show_list.append(self.show_name)

    def __str__(self):
        return f'Show name: {self.show_name}\nEpisodes: {self.no_of_episodes}\nGenre: {", ".join(self.genre)}'

    @classmethod
    def printDetails(cls):
        print(f'Total number of shows: {cls.total_shows}')
        for i in cls.show_list:
            print(i)

s1 = Netflix("Wednesday",["Mystery","Supernatural"],15)
print("==========1==========")
print(s1)
s2 = Netflix("Dark",["Mind-Bending","Sci-fi"])
print("==========2==========")
print(s2)
print("==========3==========")
Netflix.printDetails()
s3 = Netflix("Suits",["Comedy","Courtroom"],20)
print("==========4==========")
print(s3)
s4 = Netflix("Demon Slayer",["Anime"],22)
print("==========5==========")
print(s4)
print("==========6==========")
Netflix.printDetails()
```

```python
# Question 2

class Mobile:
    countryCodes = {"880": "Bangladesh", "966": "India", "455": "USA"}

    def __init__(self, model, simCardStatus):
        self.model = model
        self.__simCardStatus = simCardStatus
        print(f"Model {model} is manufactured.")

    def setSimCardStatus(self,status):
        self.__simCardStatus = status
        print("SIM card status updated successfully.")

    def getSimCardStatus(self):
        return self.__simCardStatus
    
    def __str__(self):
        return f'Mobile Phone Detail:\nModel: {self.model}\nSIM Card Status: {self.__simCardStatus}'
    

class Nokia(Mobile):
    def __init__(self, model, simCardStatus, balance = 0):
        super().__init__(model, simCardStatus)
        self.balance = balance
        self.dialCallHistory = []

    def __str__(self):
        return f'{super().__str__()}\nBalance: {self.balance} TK'

    def dialCall(self, number):
        if self.getSimCardStatus() == False:
            return "No SIM card available!"
        
        else:
            if self.balance <= 0:
                return "Insufficient balance!"
            
            else:

                if number[:3] not in Mobile.countryCodes.keys():
                    return "Dialing is not allowed in this region." 

                else:
                    for key, value in Mobile.countryCodes.items():

                        if number[:3] == key:
                             self.dialCallHistory.append(number)
                             return f'Dialing the number {number} to {value} region.'
                        

    def changeSIMCardStatus(self):
        self.setSimCardStatus(True)

    def rechargeSIMCard(self, num):
        self.balance += num
        print(f'Recharge successful! Current balance {self.balance} TK.')


N3110 = Nokia("N3110", False)
print("#######################################")
print(N3110)
print("1======================================")
N1100 = Nokia("N1100", True,100)
print("#######################################")
print(N1100)
print("2======================================")
print(N3110.dialCall("88017196xxxx"))
print("3======================================")
N3110.changeSIMCardStatus()
print("4======================================")
print(N3110.dialCall("88017196xxxx"))
print("5======================================")
N3110.rechargeSIMCard(200)
print("6======================================")
print(N3110.dialCall("88017196xxxx"))
print("7======================================")
print(N1100.dialCall("45617196xxxx"))
print("8======================================")
print(N1100.dialCall("45517196xxxx"))
print(N1100.dialCall("96617196xxxx"))
print("9======================================")
print(f"Dial call history for {N1100.model}: {N1100.dialCallHistory}")
print(f"Dial call history for {N3110.model}: {N3110.dialCallHistory}")
```

```python
# Question 3

class A:
    temp = 7

    def __init__(self):
        self.y = A.temp - 3
        self.sum = self.temp + 2
        A.temp += 3

    def methodA(self, m, n, x=0):
        self.y = self.y + m + (A.temp)
        x = x + 2 + n
        self.sum = self.sum + x + self.temp
        print(x, self.y, self.sum)
 
class B(A):
    temp = 1
 
    def __init__(self, obj=None):
        super().__init__()
        self.temp = self.temp + B.temp
        self.sum = 3 + B.temp + A.temp
        if obj != None:
            obj.methodB(3, 6)
    
        else:
            self.methodB(1, 4)
 
    def methodB(self, m, n):
        y = self.temp + self.y + n
        B.temp = m + self.y + n
        self.methodA(n, m)
        self.sum = self.y + y + A.temp
        print(self.temp , y, self.sum)

b1 = B()
b2 = B(b1)
```
