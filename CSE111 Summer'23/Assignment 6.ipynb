```python
#Task 1

class Movie:
    def __init__(self, name, cat, time):
        self.movie_name = name
        self.movie_catagory = cat
        self.movie_runtime = time

    def movieInfo(self):
        return f'Movie Name: {self.movie_name}\nMovie Genre: {self.movie_catagory}\nMovie Duration: {self.movie_runtime} Minutes'
    
    @classmethod
    def createMovie_fromString(cls, info):
        name, cat, time = info.split("-")
        obj = cls(name, cat, int(time))
        return obj

class StarCinema:
    all_branch_info = {}

    def __init__(self, name):
        self.branch_name = name
        self.movie_list = []
        print(f'Welcome to the {self.branch_name} branch of StarCinema.')

    def addMovies(self, *movies):
        for i in movies:

            if i not in self.movie_list:
                self.movie_list.append(i)
                print(f"{i.movie_name} added to the {self.branch_name} branch.")

            else:
                print(f'{i.movie_name} already exist in this branch.')

        StarCinema.all_branch_info[self.branch_name] = self.movie_list
    
    def removeMovie(self, movie):
        if movie in self.movie_list:
            self.movie_list.remove(movie)
            StarCinema.all_branch_info[self.branch_name] = self.movie_list
            print(f'{movie.movie_name} removed from {self.branch_name} branch.')
        
        else:
            print(f"{movie.movie_name} dosen't exist in this branch.")

    @classmethod
    def check(cls, movie):
        flag = False
        for key,value in cls.all_branch_info.items():

            for i in value:

                if movie == i.movie_name:
                    print(f'{movie} is being streamed at {key} branch.')
                    flag = True
                
        if flag:
            for movie_list in cls.all_branch_info.values():

                for i in movie_list:

                    if i.movie_name == movie:
                        print(f'It is of {i.movie_catagory} genre and is {i.movie_runtime} minutes long.')

                break

        else:
            print(f'The movie {movie} is not being streamed anywhere right now.')


    @classmethod
    def showAllBranchInfo(cls):
        for key,value in cls.all_branch_info.items():
            print(f'Branch Name: {key}')
            print("-------------------")
            count = 1

            for i in value:
                print(f'Movie No.: {count}')
                print(i.movieInfo())
                count +=1
                print("-------------------")


movie1 = Movie("Oppenheimer", "Biographical Drama", 180)
movie2 = Movie.createMovie_fromString("Spare Parts-Drama-115")
movie3 = Movie("Top Gun: Maverick", "Action", 131)
movie4 = Movie("Mission: Impossible - Dead Reckoning Part One", "Action", 163)
movie5 = Movie.createMovie_fromString("13 Hours: The Secret Soldiers of Benghaz-Action-144")
print("==========================================")
branch1 = StarCinema("Uttara")
print("==========================================")
branch1.addMovies(movie1, movie4)
branch1.addMovies(movie5, movie4)
print("==========================================")
StarCinema.showAllBranchInfo()
print("==========================================")
branch1.removeMovie(movie5)
print("==========================================")
StarCinema.showAllBranchInfo()
print("==========================================")
branch2 = StarCinema("Mohakhali")
print("==========================================")
branch2.addMovies(movie2, movie5)
print("==========================================")
StarCinema.showAllBranchInfo()
print("==========================================")
branch2.addMovies(movie3)
branch2.removeMovie(movie5)
branch1.addMovies(movie2)
print("==========================================")
StarCinema.check("Spare Parts")
print("==========================================")
branch1.removeMovie(movie2)
print("==========================================")
StarCinema.check("13 Hours: The Secret Soldiers of Benghaz")
StarCinema.check("Oppenheimer")
print("==========================================")
StarCinema.showAllBranchInfo()
print("==========================================")
branch3 = StarCinema("Mirpur")
print("==========================================")
branch3.addMovies(movie5)
print("==========================================")
StarCinema.showAllBranchInfo()
print("==========================================")
```
