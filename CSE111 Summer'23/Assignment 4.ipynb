```python
#Task 1

class HackathonTeam:
    def __init__(self,name, a1, a2 = None, a3 = None):
        self.team = name
        self.participant_1 = a1
        self.participant_2 = a2
        self.participant_3 = a3

    def information(self):
        print(f'Team name: {self.team}')
        print("Participants : ")
        print(self.participant_1)

        if self.participant_2 != None:
            print(self.participant_2)

        if self.participant_3 != None:
            print(self.participant_3)


team_1 = HackathonTeam("Atlantean", "Aquaman")
team_1.information()


print("=================")


team_2 = HackathonTeam("Avengers", "Ironman", "Thor", "Hulk")
team_2.information()


print("=================")


team_3 = HackathonTeam("X-Men", "Storm", "Mystique")
team_3.information()
```

```python
#Task 2

global menu 
menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}

class Foodie:

    def __init__(self, name):
        self.name = name
        self.items = []
        self.total = 0

    def order(self, *items):
        for i in items:
            item, quantity = i.split("-")
            self.items.append(item)
            unit_price = menu[item]
            total_price = unit_price * int(quantity)
            self.total += total_price
            print(f'Ordered - {item}, quantity - {quantity}, price(per Unit) - {unit_price}.')
            print(f'Total Price - {total_price}')
        
    def pay_tips(self, num = 0):
        self.total += num
        if num == 0:
            print("No tips to the waiter.")
        else:
            print(f'Gives {num}/- tips to the waiter')

    def show_orders(self):
        return (
                f'{self.name} has {len(self.items)} item(s) in the cart.\n' 
                f'Items: {self.items}\n'f'Total spent: {self.total}'
                )
    
f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
#Task 3

class Farmer:
    def __init__(self, name = None):
        self.name = name 
        self.storage = {"crops": [], "fishes": []}
        self.crop_count = 0
        self.fish_count = 0

        if name == None:
            print(f'Welcome to your farm!')

        elif type(name) == int:
            print(f'Welcome to your farm. Your farm ID is {self.name}!')

        else:
            print(f'Welcome to your farm, {self.name}!')


    def addCrops(self, *crops):
        count = 0
        for i in crops:
            self.storage["crops"].append(i)
            count += 1
        self.crop_count = len(self.storage["crops"])

        if count == 0:
            print("No crop added")

        else:
            print(f'{count} crops(s) added.')


    def addFishes(self, *fishes):
        count = 0
        for i in fishes:
            self.storage["fishes"].append(i)
            count += 1
        self.fish_count = len(self.storage["fishes"])

        if count == 0:
            print("No fish added.")

        else:
            print(f'{count} fish(s) added .')
            

    def showGoods(self):
        if self.crop_count == 0:
            print("You don't have any crop(s) .")

        else:
            print(f'You have {self.crop_count} crop(s) :')
            print(", ".join(self.storage["crops"]))

        if self.fish_count == 0:
            print("You don't have any fish(s) .")

        else:
            print(f'You have {self.fish_count} fish(s) :')
            print(", ".join(self.storage["fishes"]))


f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#Task 4

class UniversalStudiosUser:
    def __init__(self, name, age, ct):
        self.name = name
        self.age = age
        self.catagory = ct
        self.total_price = 0.0
        self.ride_list = {}
        print("Welcome to Universal Studios.")


    def selected_rides(self, *rides):
        for i in rides:
            ride, price = i.split("-")
            self.total_price += int(price)
            self.ride_list[ride] = int(price)
        print("Added ride(s) successfully.")


    def show_details(self):
        print("Your information: ")
        print(f'Name: {self.name}, Age: {self.age}, catagory: {self.catagory}')
        print("Selected Rides: ")

        for ride, price in self.ride_list.items():
            print(f'Ride: {ride}, Amount: {price} dollar(s)')

        if len(self.ride_list.keys()) > 3:

            if self.catagory == "Special":
                self.total_price -= self.total_price * (20/100)
                print(f"Congratulations!!! You've got a 20% discount" )
                print(f'Please pay {self.total_price} dollar(s).')
            
            else:
                print(f'Please pay {self.total_price} dollar(s).')
            
        else:
            print(f'Please pay {self.total_price} dollar(s).')


customer_1 = UniversalStudiosUser("Alice", 21, "Special")
print("--------- 1 ---------")
customer_1.selected_rides("Waterworld-100", "Accelerator-200", "DinoSoarin-50")
print("--------- 2 ---------")
customer_1.show_details()


print("=================")


customer_2 = UniversalStudiosUser("Bob", 20, "Normal")
print("--------- 3 ---------")
customer_2.selected_rides("Enchanted Airways-300", "Jurassic Park-500", "Accelerator-200", "DinoSoarin-50")
print("--------- 4 ---------")
customer_2.show_details()


print("=================")


customer_3 = UniversalStudiosUser("Mark", 15, "Special")
print("--------- 5 ---------")
customer_3.selected_rides("Transformers-450", "Jurassic Park-500", "Waterworld-100", "DinoSoarin-50")
print("--------- 6 ---------")
customer_3.show_details()
```
