```python
#Task 1

class Department:
    def __init__(self, name):
        self.dept_name = name
        self.student_list = []

    def findStudent(self, id):
        flag = False
        for i in self.student_list:

            if id == i.get_id():
                print("Student Info: ")
                print(f"Student Name: {i.name}")
                print(f'ID: {i.get_id()}')
                print(f'CGPA: {i.get_cgpa()}')
                flag = True

        if not flag:
            print("Student with this ID dosen't exist, Please give a valid ID")

    def addStudent(self, *students):
        flag = False
        for i in students:
            for j in self.student_list:

                if i.get_id() == j.get_id():
                    print("Student with the same ID already exists. Please try with another ID.")
                    flag = True

            if not flag:
                self.student_list.append(i)
                print(f'Welcome to {self.dept_name} department, {i.name}')

    def details(self):
        print(f'Department Name: {self.dept_name}')
        print(f'Number of students: {len(self.student_list)}')
        print(f'Details of the students: ')
        for i in self.student_list:
                print(f'Student Name: {i.name}, ID: {i.get_id()}, cgpa: {i.get_cgpa()}')


class Student:
    def __init__(self, name, id, cgpa):
        self.name = name
        self.__id = id
        self.__cgpa = cgpa

    def get_id(self):
        return self.__id

    def get_cgpa(self):
        return self.__cgpa

    def setId(self, id):
        self.__id = id


s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#Task 2

class Spaceship:
    def __init__(self, name, cp):
        self.ship_name = name
        self.capacity = cp
        self.cargo_list = []
        self.curr_load = 0
        self.rem_load = cp

    def load_cargo(self, name):
        if name.get_weight() <= self.rem_load:
            self.cargo_list.append(name.get_cargo_name())
            self.curr_load += name.get_weight()
            self.rem_load -= name.get_weight()
            
        else:
            print(f'Warning: Unable to load {name.get_cargo_name()} inside {self.ship_name}. Exceeds capacity by {name.get_weight() - self.rem_load}')

    def display_details(self):
        print(f'Spaceship Name: {self.ship_name}')
        print(f'Capacity: {self.capacity}')
        print(f'Current Cargo Weight: {self.curr_load}')
        print(f'Cargo: {self.cargo_list}')

class Cargo:
    def __init__(self, name, wt):
        self.__cargo_name = name
        self.__weight = wt

    def get_cargo_name(self):
        return self.__cargo_name

    def get_weight(self):
        return self.__weight

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```
