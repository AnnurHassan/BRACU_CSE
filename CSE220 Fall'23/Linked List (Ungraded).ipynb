```python
class Node:
  def __init__(self,elem,next = None):
    self.elem, self.next = elem,next
```

```python
class LinkedList:
    def __init__(self, arr):
        self.head = Node(arr[0])
        tail = self.head
        for i in range(1, len(arr)):
            newNode = Node(arr[i])
            tail.next = newNode
            tail = newNode

    
    def printLinkedList(self):
        temp = self.head
        while temp != None:
            if temp.next != None:
                print(temp.elem, end = "-->")

            else:
                print(temp.elem)

            temp = temp.next

    def countNodes(self):
        temp = self.head
        count = 0
        while temp != None:
            count += 1
            temp = temp.next

        return count
        

    def nodeAt(self, idx):
        temp = self.head
        count = 0
        node = None
        while temp != None:
            if count == idx:
                node = temp
                break
            
            temp = temp.next
            count += 1

        return node


    def elemAt(self,idx):
        temp = self.head
        count = 0
        elem = None
        while temp != None:
            if count == idx:
                elem = temp.elem
            
            temp = temp.next
            count += 1

        return elem


    def indexOf(self,value):
        temp = self.head
        count = 0
        while temp != None:
            if temp.elem == value:
                return count 
            
            temp = temp.next
            count += 1

        return -1


    def contains(self,value):
        temp = self.head
        while temp != None:
            if temp.elem == value:
                return True
            
            temp = temp.next
        
        return False


    def insert(self, value, idx):
        total_nodes = self.countNodes()
        if idx == 0:
            node = Node(value, None)
            node.next = self.head
            self.head = node

        elif idx > 0 and idx < total_nodes:
            n1 = self.nodeAt(idx - 1)
            n2 = self.nodeAt(idx)
            node = Node(value, n2)
            n1.next = node

        else:
            node = Node(value, None)
            last_node = self.nodeAt(total_nodes - 1)
            last_node.next = node


    def remove(self,idx):
        total_nodes = self.countNodes()
        if idx == 0:
            self.head = self.head.next

        elif idx > 0 and idx < total_nodes:
            n1 = self.nodeAt(idx - 1)
            node = n1.next
            n1.next = node.next


    def rotateRight(self):
        ln = self.head
        sln = self.head
        while ln.next != None:
            sln = ln
            ln = ln.next
        
        ln.next = self.head
        sln.next = None
        self.head = ln


    def rotateLeft(self):
        ln = self.head
        new_head = self.head.next
        while ln.next != None:
            ln = ln.next

        ln.next = self.head
        self.head.next = None
        self.head = new_head


    def reverse(self):
        new_head = None
        temp = self.head
        while temp != None:
            secondNode = temp.next
            temp.next = new_head
            new_head = temp
            temp = secondNode

        self.head = new_head
        
        


print('Create Linked List')
l1 = LinkedList(np.array([10,34,21,6,-2]))
l1.printLinkedList()
print()

print('Count Nodes of Linked List')
print(l1.countNodes())
print()

print('Find out the node at a certain index')
n = l1.nodeAt(3)
print('Node at index 3:', end = ' ')
print(n.elem if n!=None else 'Invalid Index')
print('Corner Cases:')
print('Node at index -1:', end = ' ')
n = l1.nodeAt(-1)
print(n.elem if n!=None else 'Invalid Index')
print('Node at index 5:', end = ' ')
n = l1.nodeAt(5)
print(n.elem if n!=None else 'Invalid Index')
print()

print('Find out the element at a certain index')
print('Element at index 1:', end = ' ')
print(l1.elemAt(1))
print('Corner Cases:')
print('Element at index -2:', end = ' ')
print(l1.elemAt(-2))
print('Element at index 6:', end = ' ')
print(l1.elemAt(6))
print()

print('Find out the index of an element')
print('Index of element -2', end = ' ')
print(l1.indexOf(-2))
print('Corner Cases:')
print('Index of element 23', end = ' ')
print(l1.indexOf(23))
print()

print('Find out the if linked list contains an element')
print('Linked List contains -2:', end = ' ')
print(l1.contains(-2))
print('Corner Cases:')
print('Linked List contains 3:', end = ' ')
print(l1.contains(3))
print()

print('Insert element in certain index')
print('Insert 41 at index 2:', end = ' ')
l1.insert(41,2)
l1.printLinkedList()
print('Corner Cases:')
print('Insert 25 at index 0:', end = ' ')
l1.insert(25,0)
l1.printLinkedList()
print('Insert 15 at index 7:', end = ' ')
l1.insert(15,7)
l1.printLinkedList()
print('Insert 27 at index -1:', end = ' ')
l1.insert(27,-1)
l1.printLinkedList()
print()

print('Delete node from a certain index')
print('Delete from index 5:', end = ' ')
l1.remove(5)
l1.printLinkedList()
print('Corner Cases:')
print('Delete from index 0:', end = ' ')
l1.remove(0)
l1.printLinkedList()
print('Delete from index 5:', end = ' ')
l1.remove(5)
l1.printLinkedList()
print('Delete from index -1:', end = ' ')
l1.remove(-1)
l1.printLinkedList()
print()

print('Right Rotate')
l1.rotateRight()
l1.printLinkedList()
print()

print('Left Rotate')
l1.rotateLeft()
l1.printLinkedList()
print()

print("Reverse")
l1.reverse()
l1.printLinkedList()
print()
```
