```python
# Waiting Room Management System

class DoublyNode:
	def __init__(self, elem, next, prev):
		self.elem = elem
		self.next = next
		self.prev = prev
		

class Patient:
	def __init__(self, id, name, age, bloodgroup):
		self.id = id
		self.name = name
		self.age = age
		self.bloodgroup = bloodgroup
		

class WRM:
	def __init__(self):
		self.wrm_dh = DoublyNode(None, None, None)
		self.wrm_dh.prev = self.wrm_dh
		self.wrm_dh.next = self.wrm_dh 
		
	
	def RegisterPatient(self, id, name, age, bloodgroup):
		patient = Patient(id, name, age, bloodgroup)
		node = DoublyNode(patient, self.wrm_dh, self.wrm_dh.prev)
		self.wrm_dh.prev.next = node
		self.wrm_dh.prev = node
    
	
	def ServerPatient(self):
		if self.wrm_dh.next == self.wrm_dh:
			print("There are no patient to serve. Waiting Room is empty.\n")

		else:		
			head = self.wrm_dh.next
			head.prev = None
			self.wrm_dh.next = head.next
			head.next.prev = self.wrm_dh
			head.next = None
			print(f'{head.elem.name} has been Serviced. \n')

	
	def CancelAll(self):
		if self.wrm_dh.next == self.wrm_dh:
			print("There are to appoinments to cancel. \n")

		else:
			self.wrm_dh.prev = self.wrm_dh
			self.wrm_dh.next = self.wrm_dh
			print("All appointments have been cancelled\n")


	def CanDoctorGoHome(self):
		if self.wrm_dh.next == self.wrm_dh:
			print("Yes, the doctor can go home.\n")
		
		else:
			print("No, there are patients waiting.\n")


	def ShowAllPatient(self):
		if self.wrm_dh.next == self.wrm_dh:
			print("There are no patients in the waiting room. \n")

		else:
			print("Patient list: ", end = " ")
			temp = self.wrm_dh.next
			while temp != self.wrm_dh:
				patient = temp.elem
				if temp.next == self.wrm_dh:
					print(patient.name)

				else:
					print(patient.name, end = ", ")

				temp = temp.next

			print("\n")
	

	def ReverseTheLine(self):
		if self.wrm_dh.next == self.wrm_dh:
			print("There is no Waiting line to be reversed.\n")

		else:	
			temp = self.wrm_dh.prev
			self.wrm_dh.prev = self.wrm_dh.next
			self.wrm_dh.next = temp
			while temp != self.wrm_dh:
				n = temp.prev
				temp.prev = temp.next
				temp.next = n
				temp = n
			
			print("Waiting Line has been reversed. \n")



def RunWrm():
	print("**Welcome To Waiting Room Management System**")
	wrm = WRM()
	while True:
		print_menu()
		opt = int(input("Plase Choose an Option: "))
		if opt == 1:
			print("Registering Patient! Please Wait ......")
			id = int(input("Enter Patient ID: "))
			name = input("Enter Patient Name: ")
			age = int(input("Enter Patient Age: "))
			bloodgroup = input("Enter Patient Bloodgroup: ")
			wrm.RegisterPatient(id, name, age, bloodgroup)
			print("Patient Registered Succefully!!!\n\n")
			
		elif opt == 2:
			wrm.ServerPatient()

		elif opt == 3:
			wrm.CancelAll()

		elif opt == 4:
			wrm.CanDoctorGoHome()

		elif opt == 5:
			wrm.ShowAllPatient()

		elif opt == 6:
			wrm.ReverseTheLine()

		elif opt == 7:
			print("Exitting...........")
			break
		
		else:
			print("Invalid option. Please choose from teh above list....  \n")


def print_menu():
	print("==Choose an option==")
	print("1. Register a Patient")
	print("2. Serve Patient")
	print("3. Cancel All Appoinments")
	print("4. Can Doctor Go Home")
	print("5. Show All Patient")
	print("6. Reverse the waiting line")
	print("7. Exit Program")
	print("==========================")
		
RunWrm()
```
