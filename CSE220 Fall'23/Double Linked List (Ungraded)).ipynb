```python
class Node:
    def __init__(self, e, n, p):
        self.element = e
        self.next = n
        self.prev = p
```

```python
class DoublyList:

    # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
    def __init__(self):
        self.dh = Node(None, None, None)
  

    #create a DHDCL from an array
    def create_list(self, l):
        temp = self.dh
        for i in range(len(l)):
            node = Node(l[i], self.dh, temp)
            temp.next = node
            temp = temp.next
            self.dh.prev = temp


    # Counts the number of Nodes in the list and return the number
    def countNode(self):
        temp = self.dh.next
        count = 0
        while temp != self.dh:
            count += 1
            temp = temp.next

        return count
    

    # prints the elements in the list forward
    def forwardprint(self):
        temp = self.dh.next
        while temp != self.dh:
            print(temp.element, end = " ")
            temp = temp.next


    # prints the elements in the list backward
    def backwardprint(self):
        temp = self.dh.prev
        while temp != self.dh:
            print(temp.element, end = " ")
            temp = temp.prev


    # returns the reference of the at the given index. For invalid index return None.
    def nodeAt(self, idx):
        temp = self.dh.next
        count = 0
        node = None
        while temp != self.dh:
            if count == idx:
                node = temp
                break
            
            count += 1
            temp = temp.next
        
        return node


    # returns the index of the containing the given element. if the element does not exist in the List, return -1.
    def indexOf(self, elem):
        temp = self.dh.next
        count = 0
        while temp != self.dh:
            if temp.element == elem:
                return count 
            
            count += 1
            temp = temp.next
        
        return -1


    # insert at specific index of the DHDCL
    def insert(self, elem, idx):
        total_nodes = self.countNode()
        if idx == 0:
            node = Node(elem, None, None)
            node.next = self.dh.next
            self.dh.next.prev = node
            self.dh.next = node
            node.prev = self.dh

        elif idx > 0 and idx < total_nodes:
            prev_node = self.nodeAt(idx - 1)
            last_node = self.nodeAt(idx)
            n = Node(elem, last_node, prev_node)
            prev_node.next = n
            last_node.prev = n

        else:
            n = Node(elem, self.dh, self.dh.prev)
            self.dh.prev.next = n
            self.dh.prev = n


   #removes the node of the specific index and returns the element of the node
    def remove(self, idx):
        temp = self.dh.next
        count = 0
        while temp != self.dh:
            if count == idx:
                next_elem = temp.next
                prev_elem = temp.prev
                prev_elem.next = next_elem
                next_elem.prev = prev_elem
                temp.prev = None
                temp.next = None
                return temp.element
            
            count += 1
            temp = temp.next
```
