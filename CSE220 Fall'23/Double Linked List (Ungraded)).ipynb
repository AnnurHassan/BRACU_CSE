```python
class Node:
    def __init__(self, e, n, p):
        self.element = e
        self.next = n
        self.prev = p
```

```python
class DoublyList:
    def __init__(self):
        self.dh = None

    def create_list(self, l):
        self.dh = Node(None, None, None)
        self.dh.next = self.dh
        self.dh.prev = self.dh
        temp = self.dh
        for i in range(len(l)):
            node = Node(l[i], self.dh, temp)
            temp.next = node 
            temp = temp.next
            self.dh.prev = temp



    def countNode(self):
        temp = self.dh.next
        count = 0
        while temp != self.dh:
            count += 1
            temp = temp.next

        return count
    

    def forwardprint(self):
        temp = self.dh.next
        while temp != self.dh:
            if temp.next == self.dh:
                print(temp.element)
            
            else:
                print(temp.element, end = " ")

            temp = temp.next

        print()


    def backwardprint(self):
        temp = self.dh.prev
        while temp != self.dh:
            if temp.prev == self.dh:
                print(temp.element)
            
            else:
                print(temp.element, end = " ")
            
            temp = temp.prev

        print()


    def nodeAt(self, idx):
        temp = self.dh.next
        count = 0
        node = None
        while temp != self.dh:
            if count == idx:
                node = temp

            temp = temp.next
            count += 1
        
        return node


    def indexOf(self, elem):
        temp = self.dh.next
        count = 0
        while temp != self.dh:
            if temp.element == elem:
                return count
            
            temp = temp.next
            count += 1
        
        return -1


    def insert(self, elem, idx):
        total_node = self.countNode()
        if idx == 0:
            node = Node(elem, self.dh.next, self.dh)
            self.dh.next.prev = node
            self.dh.next = node

        elif idx > 0 and idx < total_node:
            n1 = self.nodeAt(idx - 1)
            n2 = self.nodeAt(idx)
            node = Node(elem, n2, n1)
            n1.next = node
            n2.prev = node

        else:
            node = Node(elem, self.dh, self.dh.prev)
            self.dh.prev.next = node
            self.dh.prev = node


    def remove(self, idx):
        temp = self.dh.next
        count = 0
        while temp != self.dh:
            if count == idx:
                n = temp.next
                temp.prev.next = n
                n.prev = temp.prev
                temp.prev = None
                temp.next = None
                return temp.element
            
            temp = temp.next
            count += 1
```
