# String

```python
#Task 1

string = input("Input: ")
uppercase_count = 0
lowercase_count = 0

for i in string:
  if 97 <= ord(i) <= 122:
    lowercase_count += 1
  else:
    uppercase_count += 1

if lowercase_count >= uppercase_count:
    for i in string:
        if 97 <= ord(i) <= 122:
            print(i, end="")
        else:
            print(chr(ord(i)+32), end="")
else:
    for i in string:
        if 65 <= ord(i) <= 90:
            print(i, end="")
        else:
            print(chr(ord(i)-32), end="")
```

```python
#Task 5

string = input("Input: ").strip()
uppercase = "Uppercase character missing,"
lowercase = "Lowercase character missing,"
digit = "Digit missing,"
special = "Special character missing,"

for i in string:
    if ord(i) in range(65,91):
        uppercase = ""
    
    if ord(i) in range(97,123):
        lowercase = ""
    
    if ord(i) in range(48,58):
        digit = ""
    
    if i in "_$#@":
        special = "" 

output = ""
if uppercase != "":
    output += uppercase
if lowercase != "":
    output += lowercase
if digit != "":
    output += digit
if special != "":
    output += special
if output == "":
    print("OK")
else:
    print(output[:-1])
```

# List

```python
# Task 1

num_list_1 = []
num_list_2 = []
while True:
    num = input("Input: ").strip()
    if num == "STOP":
        break

    else:
        num_list_1.append(num)
        if num in num_list_2:
            continue
        else:
            num_list_2.append(num)

for i in num_list_2:
    count = 0
    for j in num_list_1:
        if i == j:
            count += 1
        else:
            continue
    print(f'{int(i)} - {count} times')
```

# Dictionary & Tuple

```python
#Task 1

string_1 = input("Input: ").strip().split(",")
string_2 = input("Input: ").strip().split(",")
new_dict = {}
value_list = []
for i in string_1:
    i = i.split(":")
    key, value = i
    new_dict[key] = int(value)

for i in string_2:
    i = i.split(":")
    key, value = i
    if key in new_dict.keys():
        new_dict[key] += int(value)
    else:
        new_dict[key] = int(value)

for i in new_dict.values():
    if i in value_list:
        continue
    else:
        value_list.append(i)

for i in range(len(value_list)):
    for j in range(i+1,len(value_list)):
        if value_list[i] > value_list[j]:
            rem = value_list[i]
            value_list[i] = value_list[j]
            value_list[j] = rem

print(new_dict)
print(f'Values: {tuple(value_list)}')
```

```python
#Task 3

string = input("Input: ").strip().split(",")
new_dict = {}
inverted_dict = {}
for i in string:
    i = i.split(":")
    key, value = i
    new_dict[key] = value

for key,value in new_dict.items():
    if value in inverted_dict.keys():
        inverted_dict[value].append(key)
    else:
        inverted_dict[value] = [key]
    
print(inverted_dict)
```

```python
#Task 1

def BMI(height, weight):
    height = height / 100
    height = height * height
    bmi_score = round((weight / height), 1)
    if bmi_score < 18.5:
        condition = "Underweight"
    elif 18.5 <= bmi_score <= 24.9:
        condition = "Normal"
    elif 25 <= bmi_score <= 30:
        condition = "Overweight"
    elif bmi_score > 30:
        condition = "Obese"
    return (f'Score is {bmi_score}. You are {condition}')

print(BMI(175,96))
```

```python
#Task 3

def calculate_price(item, location = "Mohakhali"):
    price_list = {"BBQ Chicken Cheese Burger": 250, "Beef Burger": 170, "Naga Drums": 200}
    price = price_list[item]
    tax = price * (8/100)
    if location != "Mohakhali":
        total_price = price + tax + 60
        return total_price
    else:
        total_price = price + tax + 40
        return total_price

print(calculate_price("Beef Burger"))
```

```python
#Task 4

def replace_domain(email, new_domain, old_domain = ""):
    email_list = (email.split("@"))
    if old_domain == "":
        if email_list[-1] == new_domain:
            new_email = email
            return(f'Unchanged: {new_email}')
        else:
            new_email = email_list[0] + '@' + new_domain
            return(f'Changed: {new_email}')

    elif email_list[-1] == old_domain:
        new_email = email_list[0] + '@' + new_domain
        return(f'Changed: {new_email}')

print(replace_domain('alice@sheba.xyz', 'sheba.xyz'))
```

```python
#Task 5

def check_palindrome(string):
    new_string = ""
    for i in string:
        if i == " ":
            continue
        else:
            new_string += i

    if new_string == new_string[::-1]:
        return "Palindrome"
    else:
        return "Not a palindrome"

print(check_palindrome("nurses run"))
```

```python
#Task 6

def capitalize(string):
    if 97 <= ord(string[0]) <= 122:
        new_string = chr(ord(string[0])-32)
    else:
        new_string = string[0]
    i = 1
    while i <= len(string):
        if i == len(string)-1:
            new_string += string[i]
            break
        else:
            if string[i] in ".!?":
                if 97 <= ord(string[i+2]) <= 122:
                    new_string += ". " + chr(ord(string[i+2])-32)
                else:
                    new_string += ". " + string[i+2]
                i += 3
            elif string[i] == " i ":
                new_string += chr(ord(string[i])-32)
                i += 3
            else:
                new_string += string[i]
                i += 1
    return new_string

print(capitalize("my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. Do you know my pet dog's name? I love my pet very much."))
```
